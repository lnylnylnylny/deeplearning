```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

# 크롬 드라이버 로드
driver = webdriver.Chrome()

# 메가커피 메뉴 페이지
url = "http://mega-mgccoffee.com/menu/?menu_category1=1&menu_category2=1"
driver.get(url)

time.sleep(3)  # 로딩 대기

soup = BeautifulSoup(driver.page_source, "html.parser")

menus = soup.select("#menu_list > li")

title = []
image = []

for menu in menus:
    try:
        t = menu.select_one(".cont_text_title").text.strip()
    except:
        t = "null"

    try:
        img = menu.select_one(".cont_gallery_list_img img")["src"]
    except:
        img = "null"

    title.append(t)
    image.append(img)

df = pd.DataFrame({
    "title": title,
    "image": image
})

print(df)


driver.quit()
```
