```python
import sklearn
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier # 의사결정나무 학습 예측 수행
from sklearn.model_selection import train_test_split # 데이터 세트 학습데이터와 테스트 데이터로 분리

iris = load_iris() # 붓꽃 데이터 삽입
iris_data = iris.data

iris_label = iris.target
print('iris target 값: ', iris_label)
print('iris target 명: ', iris.target_names)

iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head()
```

```python
# 학습용 데이터와 테스트용 데이터 분리

x_train, x_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=10)
```

```python
print(y_train)
```

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=10)

# 학습 수행
dt_clf.fit(x_train, y_train)

# 학습이 완료된 의사결정나무 객체에서 테스트 데이터 세트로 예측 수행
pred = dt_clf.predict(x_test)
```

```python
# 정확도 측정
from sklearn.metrics import accuracy_score
print('예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
```
